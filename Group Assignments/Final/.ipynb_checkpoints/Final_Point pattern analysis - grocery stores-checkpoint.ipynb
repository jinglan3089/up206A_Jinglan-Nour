# Week 8: Point Pattern Analysis - Grocery Stores

```python
# the usuals
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import contextily as ctx

# new for data viz
import seaborn as sns

# to explore point patterns
from pointpats import centrography
from matplotlib.patches import Ellipse
import numpy
```

```python
gdf_store = gpd.read_file ('GroceryStores.geojson')
```

```python
gdf_store.info()
```

```python
gdf_store.head()
```

```python
gdf_store=gdf_store[['Store_Name','Store_Chain','Zip_Code','City','geometry']]
```

```python
gdf_store.Store_Chain.value_counts()
```

```python
top_10_stores = gdf_store.Store_Chain.value_counts().head(10).index.tolist()
top_10_stores
```

```python
# store variable
Store_Chain = 'Ralphs'

# map
ax = gdf_store[gdf_store.Store_Chain == Store_Chain].plot(figsize=(9,9), markersize=1)

# title (including count)
ax.set_title(Store_Chain + ' (' + str(gdf_store[gdf_store.Store_Chain == Store_Chain].shape[0]) + ' stores)')

# axis
ax.axis('off')

# basemap
ctx.add_basemap(ax,
                crs='epsg:3857', # surprise! You can change the crs here!
                source=ctx.providers.CartoDB.Positron)
```

```python
# add a basemap to jointplots
# scatterplot with edges
sns.set_style('white')
g = sns.jointplot(data = gdf_store,
                  x='x', 
                  y='y',
                  s=10,
                  height=8 # make the map bigger
                 )

# Clean axes (optional)
# g.ax_joint.set_axis_off()

# add basemap
ctx.add_basemap(g.ax_joint,
                crs='epsg:3857')
```

```python
# Set up figure and axis
f, ax = plt.subplots(figsize=(12, 9))

# Generate and add hexbin with 50 hexagons in each 
# dimension, half transparency, ommitting grids with no crime,
# and using the viridis colormap
hb = ax.hexbin(
    x = gdf_store['x'], 
    y = gdf_store['y'],
    gridsize=50, 
    linewidths=1,
    alpha=0.5, 
    mincnt=1, # don't show zero
    cmap='viridis'
)

# Add basemap
ctx.add_basemap(
    ax, 
    crs='epsg:3857',
    source=ctx.providers.CartoDB.Positron
)

# Add colorbar
plt.colorbar(hb)

# Remove axes
ax.axis('off')
```

```python
# create a subset of the data
gdf_store_high = gdf_store[gdf_store.Store_Chain.isin(['Trader_joes','Whole_Foods','Sprouts'])]
```

```python
# color code by race
g = sns.jointplot(data = gdf_store_high,
                  x='x', 
                  y='y',
                  hue='Store_Chain',
                  s=10,
                  height=10 # make the map bigger
                 )

# Clean axes
g.ax_joint.set_axis_off()

# add basemap
ctx.add_basemap(g.ax_joint,
                crs='epsg:3857')
```
